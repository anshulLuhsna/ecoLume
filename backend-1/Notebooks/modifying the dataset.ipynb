```python
df = pd.read_csv("D:\hackathon\Datasets\electrical_appliance_consumption.csv")
df.head()
```

```python
# Convert 'year' column to datetime format
df['year'] = pd.to_datetime(df['year'], format='%Y')

# Add 2 years to the 'year' column
df['year'] = df['year'] + pd.DateOffset(years=2)

# Convert 'year' column back to integer format
df['year'] = df['year'].dt.year.astype(int)

# Export the DataFrame to a new CSV file with the 'year' column in 'YYYY' format
df.to_csv('adjusted_years.csv', index=False)

print(df)
```
